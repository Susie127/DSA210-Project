```python
import pandas as pd

df = pd.read_csv("student-por.csv")
df.head()
```

```python
import matplotlib.pyplot as plt

plt.hist(df["G3"])
plt.title("Final Grades Distribution")
plt.xlabel("Grade")
plt.ylabel("Count")
plt.show()
```

Students who study more tend to have higher grades.
Higher absences are associated with lower grades.

```python
import seaborn as sns

sns.boxplot(x="studytime", y="G3", data=df)
plt.title("Study Time vs Final Grade")
plt.show()
```

Students who study more tend to have higher grades.
Higher absences are associated with lower grades.

```python
sns.scatterplot(x="absences", y="G3", data=df)
plt.title("Absences vs Final Grade")
plt.show()
```

Students who study more tend to have higher grades.
Higher absences are associated with lower grades.

```python
from scipy.stats import pearsonr

corr, p = pearsonr(df["studytime"], df["G3"])
print("Studytime correlation:", corr, "p:", p)

corr, p = pearsonr(df["absences"], df["G3"])
print("Absences correlation:", corr, "p:", p)
```

The correlation results show that study time has a positive relationship with final grades, while absences have a negative relationship with grades. Both relationships are statistically significant based on the p-values.
